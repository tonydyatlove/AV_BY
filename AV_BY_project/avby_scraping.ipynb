```python
data_2010 = []

headers = {
    'Accept': '*/*',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
    }

for p in range(1, 96):
    print(p)
    
    url = f'https://cars.av.by/filter?year[min]=2010&year[max]=2010&condition[0]=1&condition[1]=2&page={p}'
    req = requests.get(url, headers=headers)
    sleep(2)
    soup = BeautifulSoup(req.text, 'lxml')
    
    cars = soup.findAll('div', class_='listing-item listing-item--top')
    cars2 = soup.findAll('div', class_='listing-item listing-item--color listing-item--top')
    cars3 = soup.findAll('div', class_='listing-item')
    cars.extend(cars2)
    cars.extend(cars3)
    
    print(f'Машин на странице: {len(cars)}')
    
    for car in cars:
        try:
            image = car.find('div', class_='listing-item__photo').find('img')['data-src']
        except:
            image = '-'
        try:
            car_name = car.find('span', class_='link-text').text
        except:
            car_name = '-'
        try:
            car_info = car.find('div', class_='listing-item__params').text
        except:
            car_info = '-'
        try:
            price_in_byn = car.find('div', class_='listing-item__price').text
        except:
            price_in_byn = '-'
        try:
            price_in_usd = car.find('div', class_='listing-item__priceusd').text
        except:
            price_in_usd = '-'
        try:
            city = car.find('div', class_='listing-item__location').text
        except: 
            city = '-'
        try:
            date = car.find('div', class_='listing-item__date').text
        except:
            date = '-'
            
        data_2010.append([image, car_name, car_info, price_in_byn,
                     price_in_usd, city, date])
```

```python
header = ['image', 'car_name', 'car_info', 'price_in_byn', 'price_in_usd', 'city', 'date']
```

```python
df_2010 = pd.DataFrame(data_2010, columns=header)
df_2010.to_csv('data_2010.csv', sep=';', encoding='utf8')
```
